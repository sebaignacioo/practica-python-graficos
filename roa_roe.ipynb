Generamos el gráfico compartido para ROA y ROE. Se realizan los siguiente pasos:

1. Se configuran los **estílos** de los gráficos. Estos estílos **se mantendrán hasta el final**, o hasta que se cambien de nuevo.
2. Se importa el archivo `roa-roe.csv`, en la raíz del proyecto.
3. Se generan los ejes, títulos y etiquetas.
4. Se guarda el archivo PNG en la carpeta, con el nombre `roa-roe.png`.
5. Se muestra el gráfico en el Jupyter Notebook.

```python
plt.rcParams['font.size'] = 18
plt.rcParams['figure.figsize'] = [15, 10]
plt.style.use('Solarize_Light2')

headers = ['Trimestre', 'ROA administración anterior', 'ROE administración anterior', 'ROA administración actual', 'ROE administración actual']

df = pd.read_csv('roa-roe.csv', sep=';', usecols=headers)

df.set_index('Trimestre').plot(linewidth=4, linestyle='-', marker='o', markersize=12, legend=False)
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))
plt.legend(loc='lower right', fontsize=15)

plt.title('Comportamiento de ROA y ROE', fontsize=25, pad=40)
plt.xlabel('Trimestre', fontsize=20, labelpad=20)
plt.ylabel('Porcentaje', fontsize=20, labelpad=20)

plt.savefig('roa-roe.png', dpi=300, bbox_inches='tight', pad_inches=0.75)
plt.show()
```
